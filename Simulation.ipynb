```python
import math
def latlng_to_xypixel(lat, lng):
    xpixel = math.floor(138.348 * (lng + 97.5) * math.cos(math.radians(lat)))
    ypixel = math.floor(138.348 * (lat - 37.0))
    return xpixel, ypixel
```

```python
lst_vehicle = []
lst_kiosk_pixels = []
lst_kiosk = []
lst_kiosk_dict_animation = []
kiosk_id = 1
for kioskNumber in range(len(data['kiosks'])):
    kioskSplit = re.split('\n|: |\(|\)|,', data['kiosks'][kioskNumber]['msg'])
    name, lat, lng = kioskSplit[1], float(kioskSplit[5]), float(kioskSplit[6])
    xpixel, ypixel = latlng_to_xypixel(lat, lng)
    lst_kiosk_pixels.append((xpixel, ypixel))
    new_kiosk = Kiosk(kiosk_id, name, lat, lng, xpixel, ypixel)
    lst_kiosk.append(new_kiosk)
    lst_kiosk_dict_animation.append({'coordinates': [lng, lat], 'msg': str(new_kiosk)})
    kiosk_id += 1
# lst_kiosk_dict_animation is what makes the kiosks icon appear
# Let's edit what the message(msg) says, so that it provides more information
```

```python
# Copied directly from Gabe's code, to reverse engineer
lst_all_passengers_within_ODD = []
# This range used to be list of unzipped files, but I don't know how many that is
for person in range(0, 1):
    curr_lst_pax = [Passenger(personID, lat, lng, dest_lat, dest_lng, oxcoord, oycoord, dxcoord, dycoord, odeparturetime, max_circuity) for personID, lat, lng, dest_lat, dest_lng, oxcoord, oycoord, dxcoord, dycoord, odeparturetime, max_circuity in \
    zip(df['Person ID'], df['OLat'], df['OLon'], df['DLon'], df['DLon'], df['OXCoord'], df['OYCoord'], df['DXCoord'], df['DYCoord'], df['ODepartureTime'], df['max_circuity']) if (oxcoord, oycoord) in lst_kiosk_pixels and (dxcoord, dycoord) in lst_kiosk_pixels and (oxcoord, oycoord) != (dxcoord, dycoord)]
    lst_all_passengers_within_ODD.extend(curr_lst_pax)
# Need lst_kiosk_pixels
print(lst_all_passengers_within_ODD[0])
# Running it for 89000 things is way too many, takes probably like 30 minutes
# But lst_all_passengers_within_ODD is good :)
```

```python
# Still trying to get the dispatcher function to run
# lst_all_passengers is needed, so this cell is defining it
from random import sample, random, randrange
df_departure_times = pd.DataFrame([pax.getDepartureTime() for pax in lst_all_passengers_within_ODD], columns=["DepartureTime"])
print("Num Pax", len(lst_all_passengers_within_ODD))

# Some modesplit stuff
TIME_STEP = 10
modesplit = 0.10
bins = range(0, max(df_departure_times['DepartureTime'])+TIME_STEP, TIME_STEP)
times = pd.Series(df_departure_times['DepartureTime'])
groups = pd.cut(times, bins=bins)
timestep_departure_counts = groups.value_counts(sort=False).to_list()
max_timestep_count = max(timestep_departure_counts)
modesplits_by_timestep = []
for count in timestep_departure_counts:
    if count == 0:
        modesplits_by_timestep.append(0)
    else:
        modesplits_by_timestep.append(min(1, (modesplit * max_timestep_count) / count))

print(len(groups), len(modesplits_by_timestep))
    
lst_all_passengers = []
for pax in lst_all_passengers_within_ODD:
    modesplit = modesplits_by_timestep[pax.getDepartureTime()//TIME_STEP]
    if random() <= modesplit:
        lst_all_passengers.append(pax)
        
# Sort all passengers by departure time from origin
lst_all_passengers.sort(key=lambda pax:pax.odeparturetime)
```
